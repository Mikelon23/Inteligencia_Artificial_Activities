```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Entradas: [m², habitaciones, baños]
X = np.array([
    [60, 2, 1],
    [80, 3, 1],
    [100, 3, 2],
    [120, 4, 2],
    [150, 4, 3],
    [200, 5, 3],
    [250, 6, 4]
], dtype=float)

# Precios (miles de USD)
y = np.array([
    [90],
    [120],
    [150],
    [180],
    [230],
    [300],
    [370]
], dtype=float)
```

```python
model = keras.Sequential([
    layers.Dense(8, input_shape=(3,), activation='relu'),  # capa oculta
        layers.Dense(4, activation='relu'),
    layers.Dense(1)  # capa de salida (precio)
])
```

```python
model.compile(optimizer='adam', loss='mean_squared_error')
```

```python
history = model.fit(X, y, epochs=1000, verbose=0)
```

```python
plt.plot(history.history['loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss (MSE)')
plt.title('Evolución del error')
plt.show()
```

```python
X_test = np.array([
    [90, 3, 1],
    [160, 4, 3],
    [220, 5, 4]
], dtype=float)

predictions = model.predict(X_test)

for features, price in zip(X_test, predictions):
    print(f"Casa de {features[0]} m², {features[1]} habs, {features[2]} baños → "
        f"precio estimado: {price[0]:.1f} mil USD")
```

**Visualización 3D del modelo entrenado (m² vs. precio)**

```python
from mpl_toolkits.mplot3d import Axes3D

# Generar un rango de m²
m2_range = np.linspace(50, 250, 50)

# Fijar valores medios de habitaciones y baños
habitaciones_fijas = 3
baños_fijos = 2

# Crear entradas de prueba combinadas
X_pred = np.array([[m2, habitaciones_fijas, baños_fijos] for m2 in m2_range])
y_pred = model.predict(X_pred)

# Crear figura 3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

# Datos originales
ax.scatter(X[:,0], X[:,1], y, color='blue', label='Datos reales', s=50)

# Superficie aprendida (m² vs. precio)
ax.plot(m2_range, [habitaciones_fijas]*len(m2_range), y_pred.flatten(),
        color='red', linewidth=2, label='Modelo aprendido')

ax.set_xlabel('Tamaño (m²)')
ax.set_ylabel('Habitaciones (fijas)')
ax.set_zlabel('Precio (miles USD)')
ax.set_title('Modelo de regresión múltiple en Keras (3D)')
ax.legend()
plt.show()
```

**Superficie 3D completa (m² y habitaciones)**

```python
# Crear una malla de valores
m2 = np.linspace(50, 250, 30)
habs = np.linspace(2, 6, 30)
m2_grid, habs_grid = np.meshgrid(m2, habs)

# Fijar los baños en 2
X_surface = np.array([[m2_grid[i][j], habs_grid[i][j], 2] 
                    for i in range(len(m2)) for j in range(len(habs))])
y_surface = model.predict(X_surface)
y_surface = y_surface.reshape(m2_grid.shape)
```

```python
# Gráfica 3D
fig = plt.figure(figsize=(9,7))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(m2_grid, habs_grid, y_surface, cmap='viridis', alpha=0.7)
ax.scatter(X[:,0], X[:,1], y, color='red', label='Datos reales')

ax.set_xlabel('Tamaño (m²)')
ax.set_ylabel('Habitaciones')
ax.set_zlabel('Precio (miles USD)')
ax.set_title('Superficie aprendida por el modelo (m² y habitaciones)')
plt.legend()
plt.show()
```
