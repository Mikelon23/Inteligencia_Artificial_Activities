**Visualizar activaciones intermedias**

```python
from tensorflow.keras import models

# Crear un nuevo modelo que devuelva las salidas de las capas intermedias
layer_outputs = [layer.output for layer in model.layers[:4]]
visualizador = models.Model(inputs=model.input, outputs=layer_outputs)

# Tomar una imagen del set de prueba
imagen = x_test[0].reshape(1,28,28,1)

# Obtener activaciones
activaciones = visualizador.predict(imagen)

# Mostrar activaciones de las dos primeras capas convolucionales
layer_names = [layer.name for layer in model.layers[:4]]
for layer_name, activacion in zip(layer_names, activaciones):
    n_filtros = activacion.shape[-1]
    size = activacion.shape[1]
    display_grid = np.zeros((size, size * n_filtros))
    for i in range(n_filtros):
        canal = activacion[0, :, :, i]
        canal -= canal.mean()
        canal /= (canal.std() + 1e-5)
        canal *= 64
        canal += 128
        canal = np.clip(canal, 0, 255).astype('uint8')
        display_grid[:, i * size : (i + 1) * size] = canal
    plt.figure(figsize=(n_filtros / 2, size / 2))
    plt.title(layer_name)
    plt.grid(False)
    plt.imshow(display_grid, aspect='auto', cmap='viridis')
    plt.show()
```
