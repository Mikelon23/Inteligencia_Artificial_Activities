***Cargar datos + normalizar***

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np, matplotlib.pyplot as plt

IMG_SIZE, BATCH = 128, 32

train = keras.utils.image_dataset_from_directory(
    "cats_vs_dogs/train", image_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH
)

val = keras.utils.image_dataset_from_directory(
    "cats_vs_dogs/val", image_size=(IMG_SIZE, IMG_SIZE), batch_size=BATCH
)

norm = layers.Rescaling(1./255)
train = train.map(lambda x,y: (norm(x), y))
val   = val.map(lambda x,y: (norm(x), y))

# sanity check
xb, yb = next(iter(train))
print(xb.shape, yb.shape)  # (B, 128, 128, 3) (B,)
```
